```python
print(data["Gender"].unique())
print(data["Married"].unique())
print(data["Phone Service"].unique())
```

```python
data["Married"] = data["Married"].replace({"Yes": "Y", "No": "N"})
data["Phone Service"] = data["Phone Service"].replace({"Yes": "Y", "No": "N"})
```

```python
data["Age"] = data["Age"].astype(int)
data["Tenure in Months"] = data["Tenure in Months"].astype(int)
data["Monthly Charge"] = data["Monthly Charge"].astype(float)
data["Total Charges"] = pd.to_numeric(data["Total Charges"], errors="coerce")
```

```python
# Plot boxplots for all features
plt.figure(figsize=(20, 15))
data.boxplot(rot=90)
plt.title("Boxplot for all features")
plt.show()
```

```python
outliers = [
    "Number of Dependents",
    "Number of Referrals",
    "Avg Monthly Long Distance Charges",
    "Avg Monthly GB Download",
    "Total Charges",
    "Total Long Distance Charges",
    "Total Revenue",
    "Total Refunds",
    "Total Extra Data Charges",
]
```

```python
for feature in outliers:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data = data[(data[feature] >= lower) & (data[feature] <= upper)]
```
